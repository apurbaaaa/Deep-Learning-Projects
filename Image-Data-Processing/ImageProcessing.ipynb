### Resizing the image ###


```python
#using pillow
from PIL import Image
```

```python
img = Image.open('./puppy-care-guide-for-new-parents.jpg')
```

```python
img_resized = img.resize((200, 200))
```

```python
img_resized.save('resized_image.jpg')
```

```python
img_res = mpimg.imread('./resized_image.jpg')
```

```python
img_res
```

```python
plt.imshow(img_res)
```

### Greyscaling using OpenCV

```python
import cv2
```

```python
img = cv2.imread('./puppy-care-guide-for-new-parents.jpg')
```

```python
grayscaled_image = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
grayscaled_image
```

```python
plt.imshow(grayscaled_image)
```

```python
cv2.imwrite('dog_grayscale_image.jpg', grayscaled_image)
```
